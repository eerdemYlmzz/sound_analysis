```python
import librosa as lb
import  numpy as np
import soundfile as sf
import matplotlib.pyplot as plt 
from pathlib import Path
```

```python
def plot_spectogram_and_save(signal,sample_rate,output_path:Path):
    stft = lb.stft(signal)
    spectogram = np.abs(stft)
    spectogram_db=lb.amplitude_to_db(spectogram)
    plt.figure(figsize=(10,4))
    
    img=lb.display.specshow(spectogram_db,y_axis="log",x_axis="time",
    sr=sample_rate,cmap="inferno")
    plt.colorbar(img,format="%+2.f dBFS")
    reversing=lb.istft(spectogram)
    #spectogramı sese çeviriyoruz.
    sf.write("yeni_ses.mp3",reversing,sample_rate)
    
    


def main():
    signal,sample_rate=sf.read(Path("data") / r"C:\Users\ASUS\Downloads\çokkolay.mp3")
    print(f"sample rate {sample_rate} , signal  {signal}" ) 
    plot_spectogram_and_save(signal,sample_rate,Path("img") / "spectogram.png")


if __name__ =="__main__":
    main()
```
